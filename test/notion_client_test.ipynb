```python
import requests
from pprint import pprint


class NotionAPI:
    def __init__(self):
        self.base_url = 'https://api.notion.com/v1'
        
        """
        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  # sesquis's notion
        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  # 
        """
        self.api_key = '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'
        self.database_id = '18584e0bd49f80788551d551654938d6'

        self.headers = {
            'Authorization': 'Bearer ' + self.api_key,
            'Notion-Version': '2022-06-28'
        }
    
    def request_get(self, endpoint):
        url = self.base_url + endpoint
        response = requests.get(url, headers=self.headers)
        return response.json()
    
    def request_post(self, endpoint, json):
        url = self.base_url + endpoint
        response = requests.post(url, headers={**self.headers, 'Content-Type': 'application/json'}, json=json)
        return response.json()
    
    def get_idea_list(self):
        request_json = {
            "filter": {
                "property": "태그",
                "multi_select": {
                "contains": "backlog"
                }
            }
        }
        response = self.request_post('/databases/' + self.database_id + '/query', request_json)
        
        items = []
        for item in response['results']:
            items.append({
                'id': item['id'], 
                'name': item['properties']['이름']['title'][0]['plain_text']
            })
        
        return items
    
    def get_idea_list_string(self):
        idea_list = self.get_idea_list()
        idea_list_string = '\n'.join([item['name'] for item in idea_list])
        return idea_list_string
    
    def get_idea(self, idea_id: str):
        response = self.request_get('/blocks/' + idea_id + '/children')

        context = []
        for item in response['results']:
            rich_text = item[item['type']]['rich_text']
            if len(rich_text) > 0:
                item = rich_text[0]['plain_text']
                context.append(item)
        
        context = '\n'.join(context)
        return context

if __name__ == '__main__':
    notion_api = NotionAPI()
    idea_list = notion_api.get_idea_list()
    idea_context = notion_api.get_idea(idea_list[0]['id'])
    
    idea_list_string = notion_api.get_idea_list_string()
    print(idea_list_string)
    
```
